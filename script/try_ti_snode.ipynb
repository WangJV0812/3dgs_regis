```python
import taichi as ti

ti.init(arch=ti.cuda)
```

```python
block = ti.root.pointer(ti.ijk, 8)
pixel = block.dynamic(ti.l, 4)
x = ti.field(ti.f32)
pixel.place(x)

@ti.kernel
def fill():
    for i, j, k in ti.ndrange(3, 3, 3):
        x[i, j, k].append(k * 10 + i * 3 + j)
        
@ti.kernel
def print_activated():
    for i, j, k, l in x:
        print(f"x[{i}, {j}, {k}, {l}] = {x[i, j, k, l]}")


fill()
print_activated()
```
